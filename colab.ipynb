```python
import numpy as np
from numpy.random import randn
from matplotlib import pyplot
import matplotlib.pyplot as plt
from IPython import display
from sklearn.preprocessing import StandardScaler
import pandas as pd
import plotly.express as px
```

A diferencia del ejemplo, se generaro X como una matriz polinomal lineal de coheficientes aleatorios. 

```python
# Dimensiones
n = 1000
# Generamos X
import random

X = random.sample(range(1,10),1) * randn(n) + random.sample(range(1,10),1)
#ax+b
```

```python
float_formatter = "{:.4f}".format
np.set_printoptions(formatter={'float_kind':float_formatter})
```

```python
#Calculamos sus medidas de tendecia central y dispercion
X_column = np.resize(X,(n,1))
scaler = StandardScaler()
X_std = scaler.fit_transform(X_column)
print('media X_std = ',np.mean(X_std).round(5))
print('desviación estándar X_std = ',np.std(X_std).round(5))
print('varianza X_std = ',np.var(X_std).round(5))
```

Para obtener Y lo obtendremos igual al ejemplo en funcion de X



```python
Y =  X  + 3.5 * randn(n)
```

```python
Y_column = np.resize(Y,(n,1))
scaler = StandardScaler()
Y_std = scaler.fit_transform(Y_column)
print('media Y_std = ',np.mean(Y_std).round(5))
print('desviación estándar Y_std = ',np.std(Y_std).round(5))
print('varianza Y_std = ',np.var(Y_std).round(5))
```

```python
# Dataframe A
A = pd.DataFrame()
A[['col_1']]= pd.DataFrame(X)
A[['col_2']]= pd.DataFrame(Y)
# Dataframe A_std
A_std = pd.DataFrame()
A_std[['col_1']]= pd.DataFrame(X_std)
A_std[['col_2']]= pd.DataFrame(Y_std)
# Dataframe A_centered
A_centered = pd.DataFrame()
A_centered[['col_1']]= pd.DataFrame(X) - pd.DataFrame(X).mean()
A_centered[['col_2']]= pd.DataFrame(Y) - pd.DataFrame(Y).mean()
```

```python
import plotly.graph_objects as go
fig_1 = px.scatter(A, x="col_1", y="col_2",
                  opacity=0.3,
                   color_discrete_sequence=['blueviolet']
                  )
fig_2 = px.scatter(A_std, x="col_1", y="col_2", 
                    opacity=0.6,
                   color_discrete_sequence=['orange']
                   )
fig_3 = px.scatter(A_centered, x="col_1", y="col_2", 
                   opacity=0.3,
                   color_discrete_sequence=['gray']
                   )
fig = go.Figure(data=fig_1.data + fig_2.data + fig_3.data )
#fig = go.Figure(data=fig_2.data  )
fig.show()
#fig.write_image("data_std_centered.pdf", width=900, height=600)
```

En esta sección se realizara una comparacion de la matriz covarancia con el calculo de $A_{std}^T$$A_{std}$

```python
Matriz_covarianza = A_std.cov()
Matriz_covarianza
```

```python
from numpy.linalg import eig
auto_valores, auto_vectores = eig(Matriz_covarianza)
print('Autovalores de la matriz de covarianza de A_std = \n', auto_valores)
print('Autovectores de la matriz de covarianza de A_std = \n', auto_vectores)
```

```python
## auto_valor[1] corresponde a auto_valor[0]
auto_valores[1] * auto_vectores[0] - Matriz_covarianza.dot(auto_vectores[0])
```

```python
## auto_valor[0] corresponde a auto_valor[1]
auto_valores[0] * auto_vectores[1] - Matriz_covarianza.dot(auto_vectores[1])
```

```python
fig = px.scatter(A_std,
                 x="col_1", 
                 y="col_2", 
                 color_discrete_sequence=['orange'], 
                 marginal_x='violin', 
                 marginal_y='violin',
                 opacity=0.6
                )


arrow_1 = go.layout.Annotation(dict(
                x=  auto_valores[1] * auto_vectores[0][0], ## auto_valor[1] corresponde a auto_valor[0]
                y=  auto_valores[1] * auto_vectores[0][1],
                xref="x", yref="y",
                text="",
                showarrow=True,
                axref = "x", ayref='y',
                ax= 0,
                ay= 0,
                arrowhead = 3,
                arrowwidth=2.5,
                arrowcolor=' mediumseagreen',)
            )

arrow_2 = go.layout.Annotation(dict(
                x=  auto_valores[0] * auto_vectores[1][0], ## auto_valor[0] corresponde a auto_valor[1]
                y=  auto_valores[0] * auto_vectores[1][1],
                xref="x", yref="y",
                text="",
                showarrow=True,
                axref = "x", ayref='y',
                ax= 0,
                ay= 0,
                arrowhead = 3,
                arrowwidth=2.5,
                arrowcolor='blueviolet',)
            )


fig.update_layout(annotations= [arrow_1,arrow_2])
fig.show()
#fig.write_image("figura_matriz_covarianza.pdf", width=900, height=600)
```

Veamos:

$corr(x,y) = \frac{{\rm cov(x,y)}}{{\rm std}(x){\rm std}(y)}=\frac{\langle x_{std}, y_{std} \rangle}{n-1}$

```python
S = A_std.T.dot(A_std) ### La matriz de covarianza es S/(n-1)
S
```

```python
auto_valores, auto_vectores = eig(S)
print(auto_valores)
print(auto_vectores)
```

```python
print('singular_value_1 =', np.sqrt(auto_valores[0]).round(4))
print('singular_value_2 =', np.sqrt(auto_valores[1]).round(4))
```

Se prueba que:

$std(\sigma_i u_i) = \frac{\sigma_i}{\sqrt{n-1}}$

```python
print('std de la componente 1', S[0]/np.sqrt(n-1))  
print('std de la componente 2', S[1]/np.sqrt(n-1))  
```

*Varianza explicada y varianza total*


Cada valor singular explica una parte de la varianza total:
$$T=\frac{\sigma_1^2+\sigma_2^2}{n-1}$$

```python
pca.explained_variance_
```

```python
varianza_explicada_1 = (S[0]*S[0]/(n-1)).round(4)
varianza_explicada_2 = (S[1]*S[1]/(n-1)).round(4)
print('varianza_explicada_1 = ',varianza_explicada_1)
print('varianza_explicada_2 = ',varianza_explicada_2)
```

```python
pca.explained_variance_ratio_ 
```

Radio Varianza explicada

* radio_varianza_explicada_1 = $\frac{\sigma_1^2}{\sigma_1^2+\sigma_2^2}$
* radio_varianza_explicada_2 = $\frac{\sigma_2^2}{\sigma_1^2+\sigma_2^2}$



```python
radio_varianza_explicada_1 = (S[0]*S[0]/(S[0]*S[0]+S[1]*S[1])).round(4)
radio_varianza_explicada_2 = (S[1]*S[1]/(S[0]*S[0]+S[1]*S[1])).round(4)
print('radio_varianza_explicada_1 = ', radio_varianza_explicada_1)
print('radio-varianza_explicada_2 = ', radio_varianza_explicada_2)
```
